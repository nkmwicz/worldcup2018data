```python
events = pd.read_json('./events_World_Cupv2.json')
players = pd.read_json('./players.json')
matches = pd.read_json('./matches_World_Cup.json')
teams = pd.read_json('./teams.json')
```

```python
teams_map = teams.set_index('wyId')['name']
events['teamId'] = teams_map[events['teamId']].values
events
```

```python
players.head()
```

```python
players_map = players.set_index('wyId')[['firstName', 'lastName']]
players_map['name'] = players_map['firstName'] + ' ' + players_map['lastName']
# players_map
events['playerId'] = players_map['name'][events['playerId']].values
```

```python
events.head()
```

```python
matches.head()
matches_map = matches.set_index('wyId')['label']
events['matchId'] = matches_map[events['matchId']].values
```

```python
events.head()
```

```python
subEvents = pd.read_csv('./eventid2name.csv')
subEvents_map = subEvents.set_index('subevent')['subevent_label']
subEvents_map.loc[''] = 'N/A'
events['subEventId'] = subEvents_map[events['subEventId']].values  
# events[events['subEventId'] == '']
```

```python
events.head()
```

```python
tags = pd.read_csv('./tags2name.csv')
tags_map = tags.set_index('Tag')['Description']
tags_map
```

```python
def replace_tag_ids(tag_list):
    return [tags_map[tag['id']] for tag in tag_list]

events['tags'] = events['tags'].apply(replace_tag_ids)
```

```python
events.to_csv('./cleaned_events_world_cup2018.csv', index=False)
```
